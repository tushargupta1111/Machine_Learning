# Normal Distribution Z-Score

```python
df = pd.read_csv('weight-height.csv')
df
```

```python
df['Height'].describe()
```

```python
sns.displot(df['Height'],kde=True)
```

```python
# find boundary values
upper_boundary = df['Height'].mean() + 3*df['Height'].std()
lower_boundary = df['Height'].mean() - 3*df['Height'].std()
print(upper_boundary)
print(lower_boundary)
```

```python
df[(df['Height'] > 77.91) | (df['Height'] < 54.82)]
```

```python
sns.displot(df['Weight'],kde=True)
```

```python
# find boundary values
upper_boundary = df['Weight'].mean() + 3*df['Weight'].std()
lower_boundary = df['Weight'].mean() - 3*df['Weight'].std()
print(upper_boundary)
print(lower_boundary)
```

```python
df[(df['Weight'] > 257.76) | (df['Weight'] < 65.11)]
```

# IQR Based Filtering

```python
df
```

```python
sns.boxplot(df['Height'])
```

```python
# find IQR
Q1 = df['Height'].quantile(0.25)
Q3 = df['Height'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
upper_boundary = Q3 + 1.5*IQR
lower_boundary = Q1 - 1.5*IQR
print(upper_boundary)
print(lower_boundary)
```

```python
df[((df['Height'] > 77.67) | (df['Height'] < 55))]
```
